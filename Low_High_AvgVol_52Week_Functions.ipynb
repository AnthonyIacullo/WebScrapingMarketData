```python
tick = yf.Ticker('NVAX')
ohlc = tick.history(period='1y', interval='1D')  # pulls OHLC data from past year
df = pd.DataFrame(ohlc)  # OHLC dataframe
df['Dividends']  # deletes Dividends column
df['Stock Splits']  # deletes Stock Splits column

# delete empty rows to allow calculated data not to be messed up
df.dropna(subset=['Open'], inplace=True)

rows, columns = df.shape  # calculates the number of rows & columns... mainly use rows for calculations

rows
```

```python
df.dropna(subset=['Volume'], inplace=True)
df
```

```python
from numpy.ma.core import mean

def get_52Week_Low(dframe):
  df = dframe
  low_52Week = min(df['Low'])
  print("52 week low:", '{:.2f}'.format(low_52Week))

def get_52Week_High(dframe):
  df = dframe
  high_52Week = max(df['High'])
  print("52 week high:", '{:.2f}'.format(high_52Week))

def get_52Week_AvgVol(dframe):
  df = dframe
  avg_vol_52week =mean(df['Volume'])
  print("52 week average volume:", '{:.2f}'.format(avg_vol_52week))


get_52Week_Low(df)
get_52Week_High(df)
get_52Week_AvgVol(df)
```
